# Piedra papel o tijera

```python
# Elección de ordenador:
import random 
lista_opciones = ["piedra","papel","tijera"]

#Para ir creando nuestro marcador y de un resultado al final, tenemos que crear dos listas vacias que se vayan rellenando según quien vaya ganando.
marcador_usuario = []
marcador_ordenador = []

#Definir que hacer en cada caso, que imprimir y que añadir a los marcadores
#También se define en cada caso, el incremento del marcador según gane, pierda o empate y se va añadiendo a la lista vacía
def usuario_gana():
    print("___________________________________________________")
    print(f"Tu elegiste: {usuario_eleccion} y el ordenador eligió: {ordenador_eleccion}")
    marcador_usuario.append(1)
    marcador_ordenador.append(0)
    print(f"Usuario: {marcador_usuario} | Ordenador: {marcador_ordenador}")
    print("Ganaste! mmmmmmm pero voy con la revancha 😎")
  

def usuario_pierde():
    print("___________________________________________________")
    print(f"Tu elegiste: {usuario_eleccion} y el ordenador eligió: {ordenador_eleccion}")
    marcador_usuario.append(0)
    marcador_ordenador.append(1)
    print(f"Usuario: {marcador_usuario} | Ordenador: {marcador_ordenador}")
    print("Buuuuuu! Perdiste! 😱")
  

def usuario_empate():
    print("___________________________________________________")
    print(f"Tu elegiste: {usuario_eleccion} y el ordenador eligió: {ordenador_eleccion}")
    marcador_usuario.append(0)
    marcador_ordenador.append(0)
    print(f"Usuario: {marcador_usuario} | Ordenador: {marcador_ordenador}")
    print("Tenemos un empate! 😤")

    


#Para que solamente haya 3 oportunidades de jugar, con while metemos un límite de jugadas

limite_partidas = 3
cantidad_partidas_contador = 0 #cuenta las partidas 

#Presentación del juego
nombre_jugador = input("¿Cuál es tu nombre?: ")
print(f"Hola {nombre_jugador.upper()}!! Conóces el juego de piedra papel o tijera? \nNo?\nNo pasa nada, te explico:\nPiedra aplasta tijera, tijera corta papel y papel envuelve piedra\nListo?\nAllá que vamos!!💪🏼\n")

while cantidad_partidas_contador < limite_partidas:
    #Elección usuario
    usuario_eleccion = input("Comencemos!!!    Elige tu elemento de combate 😈 !! : piedra, papel o tijera").lower()
    #Eleccion ordenador
    ordenador_eleccion = random.choice(lista_opciones)

        #Ahora empieza el juego
    if usuario_eleccion == "papel":
        if ordenador_eleccion == "tijera":
            usuario_pierde()
        elif ordenador_eleccion == "papel":
            usuario_empate()
        elif ordenador_eleccion == "piedra":
            usuario_gana()

    elif usuario_eleccion == "tijera":
        if ordenador_eleccion == "tijera":
            usuario_empate()
        elif ordenador_eleccion == "papel":
            usuario_gana()
        elif ordenador_eleccion == "piedra":
            usuario_pierde()

    elif usuario_eleccion == "piedra":
        if ordenador_eleccion == "tijera":
            usuario_gana()
        elif ordenador_eleccion == "papel":
            usuario_pierde()
        elif ordenador_eleccion == "piedra":
            usuario_empate()

    else:
        print("Ups! esa opción no es válida campeón! 🤪")
        cantidad_partidas_contador -= 1

    cantidad_partidas_contador += 1 #vamos sumando las partidas 
    

# Imprimir resultados finales marcador de ambos bandos:

print("___________________________________________________") 
print(f"\nMarcador:\nUsuario: {sum(marcador_usuario)}     VS     Ordenador: {sum(marcador_ordenador)}")

if sum(marcador_usuario) > sum(marcador_ordenador):
    print("\n¡¡¡FELICIDADES!!! Ganaste el juego 🥳 ")
elif sum(marcador_usuario) < sum(marcador_ordenador):
    print("\nOohhhh buena suerte para la próxima ... 💃🏼🤪")
else:
    print("\nEMPATE!! ⚔️")

```
