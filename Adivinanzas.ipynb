Juego de preguntas y respuestas: es un juego en el que se hacen preguntas sobre una variedad de temas y los jugadores deben responder correctamente para avanzar. Las preguntas pueden ser de diferentes categorías, como historia, ciencias, entretenimiento o cultura general. El objetivo del juego es responder la mayor cantidad de preguntas correctamente para ganar el juego. Este juego puede ser jugado de forma individual o en equipo y es una forma divertida de aprender y poner a prueba tus conocimientos.

jugadores: tenemos la base de respuestas y hay dos jugadores, tu y el ordenador, a ver quien gana

```python


#FUNCIONES

#Función para elegir de manera aleatoria las preguntas

import random
def eleccion_de_preguntas_random ():
    # Seleeccion de pregunta aleatoria. Utilizamos la lista de preguntas, que seran keys del diccionario

    seleccionar_pregunta = random.choice(lista_preguntas_trivia) 
    
    return seleccionar_pregunta



#Vamos a poner un tope  máximo de preguntas, para que el juego no dure mucho tiempo, sobretodo para la presentación, haremos un máximo de 6 preguntas.
limite_partidas = 6
contador_partidas = 0

# Agregamos un marcador para que jugador pueda var como va 
marcador = []

# Instrucciones del juego y bienvenida

print("¿ Te consideras un sabelotodo? mmm. Pongámoslo a prueba🧐. \n Para probar tus conocimientos y comprobar que realmente eres un sabelotodo.\n Te haremos varias preguntas de diferentes temas y si respondes de forma correcta 1 punto para ti 🤸🏽. \n Si te equivocas, tienes 0 puntos.\n Al final, de acuerdo a la cantidad de puntos que tengas podrás, o no estar, en el podio🥇🥈🥉 .  Tienes 6 oportunidades \n ¡ Empecemos !" )

while contador_partidas < limite_partidas:
        # seleccion de pregunta random. Llamamos la funcion
        pregunta = eleccion_de_preguntas_random()
        
        # Respuesta del usuario
        respuesta_usuario = input(f" Vamos a ver que tanto sabes. Escribe la respuesta a esta pregunta: {pregunta}. Intenta ser muy clara en tu respuesta").lower()

        # Ponemos un 'in' para evitar errores por una respuesta literal, asi puede encontrar cualquier coincidencia

        if respuesta_usuario in  preguntas_respuestas_trivia[pregunta].lower():
            print (f"Correcto!!! La respuesta a la pregunta {pregunta} era en efecto {preguntas_respuestas_trivia[pregunta].capitalize()}")
            marcador.append(1)
            print(f"Partidas : {contador_partidas} Puntaje: {marcador}/{limite_partidas}")
            contador_partidas +=1
        
        else:
            print(f"Ésta era un poco difícil...\n La respuesta a la pregunta: {pregunta} es: {preguntas_respuestas_trivia[pregunta].capitalize()}")
            marcador.append(0)
            contador_partidas +=1
            print(f"Partidad {contador_partidas} Puntaje: {marcador}/{limite_partidas}")


    # Imprimir resultados finales y marcador

print(f"Marcador {sum(marcador)} usuario: {marcador}/{limite_partidas}")

if sum(marcador) == 6:
        print("¡¡¡Felicidades!!! Medalla de oro para ti !🥇")
elif 4 <= sum(marcador) <=6:
        print("Muy bien! Medalla de plata para ti! 🥈 ")
elif 3 <= sum(marcador) <= 4:
        print("Muy bien! Medalla de bronce para ti! 🥉 ")
elif sum(marcador) <2:
        print("Hey!! No te desanimes, ningún deportista llegó a las olimpiadas sin antes haber entrenado duro! Seguro a la próxima te va mejor 🏋️‍♀️")
    
    
```
