# Juego de preguntas y respuestas
_____________________________

Juego de preguntas y respuestas: es un juego en el que se hacen preguntas sobre una variedad de temas y los jugadores deben responder correctamente para avanzar. Las preguntas pueden ser de diferentes categorías, como historia, ciencias, entretenimiento o cultura general. El objetivo del juego es responder la mayor cantidad de preguntas correctamente para ganar el juego. Este juego puede ser jugado de forma individual o en equipo y es una forma divertida de aprender y poner a prueba tus conocimientos.

```python
#DICCIONARIOS

# Ponemos todo en un solo diccionario. Si juntamos los diccionarios en una sola lista, no podremos acceder por key-value, ya que cada pregunta y respuesta nos queda como un valor independiente. 
# Para el juego de adivinanzas, tendremos 3 categorias: cultura general, deportes y música y para ellos con 5 preguntas cada uno.

preguntas_respuestas_trivia= {# Deporte
    "País en dónde se celebraron los juegos olímpicos 2024":"francia",
    "Dónde serán los próximos juegos olímpicos":"los angeles",
    "Cuántas veces se han celebrado los juegos olímpicos en Francia":"3", 
    "Cómo se llama el nuevo deporte celebrado por primera vez en los juegos olímpicos 2024":"breaking",
    "Cuántas veces se han celebrado los juegos olímpicos":"29",
    "Qué representan los aros de los juegos olímpicos":"continentes", 
    
    # Cultura General
    "¿Cuál es el océano más grande del mundo?": "oceano pacifico",
    "¿En qué continente se encuentra Egipto?": "africa",
    "¿Cuál es la capital de Francia?": "paris",
    "¿Qué planeta es conocido como el 'Planeta Rojo'?": "marte",
    "¿Quién escribió 'Don Quijote de la Mancha'?": "miguel de cervantes",

    # Música
    "¿Quién es conocido como el 'Rey del Pop'?": "michael jackson",   
    "¿Quién es conocido como el 'Rey del Rock and Roll'?": "elvis presley",
    "¿Qué banda británica es famosa por la canción 'Hey Jude'?": "the beatles",
    "¿Qué cantante pop es conocida como la 'Reina del Pop'?": "madonna",
    "¿Qué artista jamaiquino es considerado el mayor representante del reggae?": "bob marley",
    "¿En qué país nació el compositor Ludwig van Beethoven?":"alemania"}

lista_preguntas_trivia= list(preguntas_respuestas_trivia.keys())

nombre_jugador = input("Introduce nombre de jugador: ")

# Instrucciones del juego y bienvenida
print(f"HOLA {nombre_jugador} BIENVENIDO/A AL JUEGO DEL TRIVIAL 🎉 \n \n ¿ Te consideras un sabelotodo? mmm. Pongámoslo a prueba🧐. \n Para probar tus conocimientos y comprobar que realmente eres un sabelotodo.\n Te haremos varias preguntas de diferentes temas y si respondes de forma correcta 1 punto para ti 🤸🏽. \n Si te equivocas, tienes 0 puntos.\n Al final, de acuerdo a la cantidad de puntos que tengas podrás, o no estar, en el podio🥇🥈🥉 . Solamente puedes fallar 3 veces \n Mucha suerte 🍀 \n ¡ Empecemos ! \n" )


#Función para elegir de manera aleatoria las preguntas

import random

def eleccion_de_preguntas_random():
    # Seleeccion de pregunta aleatoria. Utilizamos la lista de preguntas, que seran keys del diccionario
    seleccionar_pregunta = random.choice(lista_preguntas_trivia)

    # Se elimina la pregunta random que ha salido de la lista para que no vuelva a salir
    lista_preguntas_trivia.remove(seleccionar_pregunta)

    return seleccionar_pregunta



#Vamos a poner un tope  máximo de preguntas, para que el juego no dure mucho tiempo, sobretodo para la presentación, haremos un máximo de 8 preguntas.
limite_partidas_acertadas = 5
limite_partidas_perdidas = 3
partidas_acertadas = 0
partidas_perdidas = 0

# Agregamos un marcador para que jugador pueda var como va 
marcador = []

# Se inicia el ciclo While
while partidas_acertadas < limite_partidas_acertadas and partidas_perdidas < limite_partidas_perdidas:

        # seleccion de pregunta random. Llamamos la funcion
        pregunta = eleccion_de_preguntas_random()

        # Respuesta del usuario
        respuesta_usuario = input(f" Vamos a ver que tanto sabes. Escribe la respuesta a esta pregunta: {pregunta}. Intenta ser muy clara en tu respuesta").lower()

        if respuesta_usuario ==  preguntas_respuestas_trivia[pregunta].lower():
                print (f"\nCorrecto!!! La respuesta a la pregunta {pregunta} es en efecto: {preguntas_respuestas_trivia[pregunta].capitalize()}")
                marcador.append(1)
                partidas_acertadas +=1
                print(f"Puntaje:\nPartidas ganadas: {partidas_acertadas}/{limite_partidas_acertadas}  -  Partidas perdidas {partidas_perdidas}/{limite_partidas_perdidas}")
                print("___________________________________________________")
        else:
                print(f"\nÉsta era un poco difícil...\nLa respuesta a la pregunta: {pregunta} es: {preguntas_respuestas_trivia[pregunta].capitalize()}")
                marcador.append(0)
                partidas_perdidas +=1
                print(f"Puntaje:\nPartidas ganadas: {partidas_acertadas}/{limite_partidas_acertadas}  -  Partidas perdidas {partidas_perdidas}/{limite_partidas_perdidas}")
                print("___________________________________________________")
 

    # Imprimir resultados finales y marcador

print(f"Partidas ganadas: {partidas_acertadas}/{limite_partidas_acertadas} - Partidas jugadas: {len(marcador)}")

if partidas_acertadas == 5:
        print("\n¡¡¡Felicidades!!! Medalla de ORO para ti !🥇")
elif 4 <= partidas_acertadas <= 5:
        print("\nMuy bien! Medalla de PLATA para ti! 🥈 ")
elif 2 <= partidas_acertadas <= 3 :
        print("\nMuy bien! Medalla de BRONCE para ti! 🥉 ")
elif partidas_acertadas < 2:
        print("\nHey!! No te desanimes, ningún deportista llegó a las olimpiadas sin antes haber entrenado duro! Seguro a la próxima te va mejor 🏋️‍♀️")
    
    
```
