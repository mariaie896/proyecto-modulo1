Las preguntas pueden ser de diferentes categorÃ­as, como historia, ciencias, entretenimiento o cultura general. El objetivo del juego es responder la mayor cantidad de preguntas correctamente para ganar el juego.
```python
#DICCIONARIOS

# Ponemos todo en un solo diccionario. Si juntamos los diccionarios en una sola lista, no podremos acceder por key-value, ya que cada pregunta y respuesta nos queda como un valor independiente. 
# Para el juego de adivinanzas, tendremos 3 categorias: cultura general, deportes y mÃºsica y para ellos con 5 preguntas cada uno.

preguntas_respuestas_trivia= {# Deporte
    "PaÃ­s en dÃ³nde se celebraron los juegos olÃ­mpicos 2024":"francia",
    "DÃ³nde serÃ¡n los prÃ³ximos juegos olÃ­mpicos":"los angeles",
    "CuÃ¡ntas veces se han celebrado los juegos olÃ­mpicos en Francia":"3", 
    "CÃ³mo se llama el nuevo deporte celebrado por primera vez en los juegos olÃ­mpicos 2024":"breaking",
    "CuÃ¡ntas veces se han celebrado los juegos olÃ­mpicos":"29",
    "QuÃ© representan los aros de los juegos olÃ­mpicos":"continentes", 
    
    # Cultura General
    "Â¿CuÃ¡l es el ocÃ©ano mÃ¡s grande del mundo?": "oceano pacifico",
    "Â¿En quÃ© continente se encuentra Egipto?": "africa",
    "Â¿CuÃ¡l es la capital de Francia?": "paris",
    "Â¿QuÃ© planeta es conocido como el 'Planeta Rojo'?": "marte",
    "Â¿QuiÃ©n escribiÃ³ 'Don Quijote de la Mancha'?": "miguel de cervantes",

    # MÃºsica
    "Â¿QuiÃ©n es conocido como el 'Rey del Pop'?": "michael jackson",   
    "Â¿QuiÃ©n es conocido como el 'Rey del Rock and Roll'?": "elvis presley",
    "Â¿QuÃ© banda britÃ¡nica es famosa por la canciÃ³n 'Hey Jude'?": "the beatles",
    "Â¿QuÃ© cantante pop es conocida como la 'Reina del Pop'?": "madonna",
    "Â¿QuÃ© artista jamaiquino es considerado el mayor representante del reggae?": "bob marley",
    "Â¿En quÃ© paÃ­s naciÃ³ el compositor Ludwig van Beethoven?":"alemania"}

lista_preguntas_trivia= list(preguntas_respuestas_trivia.keys())

nombre_jugador = input("Introduce nombre de jugador: ")

# Instrucciones del juego y bienvenida
print(f"HOLA {nombre_jugador} BIENVENIDO/A AL JUEGO DEL TRIVIAL ğŸ‰ \n \n Â¿ Te consideras un sabelotodo? mmm. PongÃ¡moslo a pruebağŸ§. \n Para probar tus conocimientos y comprobar que realmente eres un sabelotodo.\n Te haremos varias preguntas de diferentes temas y si respondes de forma correcta 1 punto para ti ğŸ¤¸ğŸ½. \n Si te equivocas, tienes 0 puntos.\n Al final, de acuerdo a la cantidad de puntos que tengas podrÃ¡s, o no estar, en el podioğŸ¥‡ğŸ¥ˆğŸ¥‰ . Solamente puedes fallar 3 veces \n Mucha suerte ğŸ€ \n Â¡ Empecemos ! \n" )


#FunciÃ³n para elegir de manera aleatoria las preguntas

import random

def eleccion_de_preguntas_random():
    # Seleeccion de pregunta aleatoria. Utilizamos la lista de preguntas, que seran keys del diccionario
    seleccionar_pregunta = random.choice(lista_preguntas_trivia)

    # Se elimina la pregunta random que ha salido de la lista para que no vuelva a salir
    lista_preguntas_trivia.remove(seleccionar_pregunta)

    return seleccionar_pregunta



#Vamos a poner un tope  mÃ¡ximo de preguntas, para que el juego no dure mucho tiempo, sobretodo para la presentaciÃ³n, haremos un mÃ¡ximo de 8 preguntas.
limite_partidas_acertadas = 5
limite_partidas_perdidas = 3
partidas_acertadas = 0
partidas_perdidas = 0

# Agregamos un marcador para que jugador pueda var como va 
marcador = []

# Se inicia el ciclo While
while partidas_acertadas < limite_partidas_acertadas and partidas_perdidas < limite_partidas_perdidas:

        # seleccion de pregunta random. Llamamos la funcion
        pregunta = eleccion_de_preguntas_random()

        # Respuesta del usuario
        respuesta_usuario = input(f" Vamos a ver que tanto sabes. Escribe la respuesta a esta pregunta: {pregunta}. Intenta ser muy clara en tu respuesta").lower()

        if respuesta_usuario ==  preguntas_respuestas_trivia[pregunta].lower():
                print (f"\nCorrecto!!! La respuesta a la pregunta {pregunta} es en efecto: {preguntas_respuestas_trivia[pregunta].capitalize()}")
                marcador.append(1)
                partidas_acertadas +=1
                print(f"Puntaje:\nPartidas ganadas: {partidas_acertadas}/{limite_partidas_acertadas}  -  Partidas perdidas {partidas_perdidas}/{limite_partidas_perdidas}")
                print("___________________________________________________")
        else:
                print(f"\nÃ‰sta era un poco difÃ­cil...\nLa respuesta a la pregunta: {pregunta} es: {preguntas_respuestas_trivia[pregunta].capitalize()}")
                marcador.append(0)
                partidas_perdidas +=1
                print(f"Puntaje:\nPartidas ganadas: {partidas_acertadas}/{limite_partidas_acertadas}  -  Partidas perdidas {partidas_perdidas}/{limite_partidas_perdidas}")
                print("___________________________________________________")
 

    # Imprimir resultados finales y marcador

print(f"Partidas ganadas: {partidas_acertadas}/{limite_partidas_acertadas} - Partidas jugadas: {len(marcador)}")

if partidas_acertadas == 5:
        print("\nÂ¡Â¡Â¡Felicidades!!! Medalla de ORO para ti !ğŸ¥‡")
elif 4 <= partidas_acertadas <= 5:
        print("\nMuy bien! Medalla de PLATA para ti! ğŸ¥ˆ ")
elif 2 <= partidas_acertadas <= 3 :
        print("\nMuy bien! Medalla de BRONCE para ti! ğŸ¥‰ ")
elif partidas_acertadas < 2:
        print("\nHey!! No te desanimes, ningÃºn deportista llegÃ³ a las olimpiadas sin antes haber entrenado duro! Seguro a la prÃ³xima te va mejor ğŸ‹ï¸â€â™€ï¸")
    
    
```
