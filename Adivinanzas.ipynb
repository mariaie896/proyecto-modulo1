Juego de preguntas y respuestas: es un juego en el que se hacen preguntas sobre una variedad de temas y los jugadores deben responder correctamente para avanzar. Las preguntas pueden ser de diferentes categorÃ­as, como historia, ciencias, entretenimiento o cultura general. El objetivo del juego es responder la mayor cantidad de preguntas correctamente para ganar el juego. Este juego puede ser jugado de forma individual o en equipo y es una forma divertida de aprender y poner a prueba tus conocimientos.

```python
#DICCIONARIOS

# Ponemos todo en un solo diccionario. Si juntamos los diccionarios en una sola lista, no podremos acceder por key-value, ya que cada pregunta y respuesta nos queda como un valor independiente. 
# Para el juego de adivinanzas, tendremos 3 categorias: cultura general, deportes y mÃºsica y para ellos con 5 preguntas cada uno.

preguntas_respuestas_trivia= {# Deporte
    "PaÃ­s en dÃ³nde se celebraron los juegos olÃ­mpicos 2024":"Francia",
    "DÃ³nde serÃ¡n los prÃ³ximos juegos olÃ­mpicos":"los angeles",
    "CuÃ¡ntas veces se han celebrado los juegos olÃ­mpicos en Francia":"3", 
    "CÃ³mo se llama el nuevo deporte celebrado por primera vez en los juegos olÃ­mpicos 2024":"breaking dance",
    "CuÃ¡ntas veces se han celebrado los juegos olÃ­mpicos":"29",
    "QuÃ© representan los aros de los juegos olÃ­mpicos":"Los 5 continentes principales", 
    
    # Cultura General
    "Â¿CuÃ¡l es el ocÃ©ano mÃ¡s grande del mundo?": "ocÃ©ano pacÃ­fico",
    "Â¿En quÃ© continente se encuentra Egipto?": "Ã¡frica",
    "Â¿CuÃ¡l es la capital de Francia?": "parÃ­s",
    "Â¿QuÃ© planeta es conocido como el 'Planeta Rojo'?": "marte",
    "Â¿QuiÃ©n escribiÃ³ 'Don Quijote de la Mancha'?": "miguel de cervantes",

    # MÃºsica
    "Â¿QuiÃ©n es conocido como el 'Rey del Pop'?": "michael jackson",   
    "Â¿QuiÃ©n es conocido como el 'Rey del Rock and Roll'?": "elvis presley",
    "Â¿QuÃ© banda britÃ¡nica es famosa por la canciÃ³n 'Hey Jude'?": "the beatles",
    "Â¿QuÃ© cantante pop es conocida como la 'Reina del Pop'?": "madonna",
    "Â¿QuÃ© artista jamaiquino es considerado el mayor representante del reggae?": "bob marley",
    "Â¿En quÃ© paÃ­s naciÃ³ el compositor Ludwig van Beethoven?":"alemania"}

lista_preguntas_trivia= list(preguntas_respuestas_trivia.keys())

nombre_jugador = input("Introduce nombre de jugador: ")

print(f"HOLA {nombre_jugador} BIENVENIDO/A AL JUEGO DEL TRIVIAL ğŸ‰ \n \n Â¿ Te consideras un sabelotodo? mmm. PongÃ¡moslo a pruebağŸ§. \n Para probar tus conocimientos y comprobar que realmente eres un sabelotodo.\n Te haremos varias preguntas de diferentes temas y si respondes de forma correcta 1 punto para ti ğŸ¤¸ğŸ½. \n Si te equivocas, tienes 0 puntos.\n Al final, de acuerdo a la cantidad de puntos que tengas podrÃ¡s, o no estar, en el podioğŸ¥‡ğŸ¥ˆğŸ¥‰ .  Tienes 6 oportunidades \n Mucha suerte ğŸ€ \n Â¡ Empecemos ! \n" )


#FunciÃ³n para elegir de manera aleatoria las preguntas

import random

def eleccion_de_preguntas_random():
    # Seleeccion de pregunta aleatoria. Utilizamos la lista de preguntas, que seran keys del diccionario
    seleccionar_pregunta = random.choice(lista_preguntas_trivia)

    # Se elimina la pregunta random que ha salido de la lista para que no vuelva a salir
    lista_preguntas_trivia.remove(seleccionar_pregunta)

    return seleccionar_pregunta



#Vamos a poner un tope  mÃ¡ximo de preguntas, para que el juego no dure mucho tiempo, sobretodo para la presentaciÃ³n, haremos un mÃ¡ximo de 6 preguntas.
limite_partidas = 6
contador_partidas = 0

# Agregamos un marcador para que jugador pueda var como va 
marcador = []

# Instrucciones del juego y bienvenida

while contador_partidas < limite_partidas:
        # seleccion de pregunta random. Llamamos la funcion
        pregunta = eleccion_de_preguntas_random()
        
        # Respuesta del usuario
        respuesta_usuario = input(f" Vamos a ver que tanto sabes. Escribe la respuesta a esta pregunta: {pregunta}. Intenta ser muy clara en tu respuesta")

        # Ponemos un 'in' para evitar errores por una respuesta literal, asi puede encontrar cualquier coincidencia

        if respuesta_usuario in  preguntas_respuestas_trivia[pregunta].lower():
            print (f"\nCorrecto!!! La respuesta a la pregunta {pregunta} era en efecto {preguntas_respuestas_trivia[pregunta].capitalize()}")
            marcador.append(1)
            print(f"Partidas : {contador_partidas} Puntaje: {marcador}/{limite_partidas}")
            print("___________________________________________________")
            contador_partidas +=1
            

        else:
            print(f"\nÃ‰sta era un poco difÃ­cil...\nLa respuesta a la pregunta: {pregunta} es: {preguntas_respuestas_trivia[pregunta].capitalize()}")
            marcador.append(0)
            contador_partidas +=1
            print(f"Partidad {contador_partidas} Puntaje: {marcador}/{limite_partidas}")
            print("___________________________________________________")

    # Imprimir resultados finales y marcador

print(f"Marcador final: {sum(marcador)} / {limite_partidas}")

if sum(marcador) == 6:
        print("\nÂ¡Â¡Â¡Felicidades!!! Medalla de ORO para ti !ğŸ¥‡")
elif 4 <= sum(marcador) <=6:
        print("\nMuy bien! Medalla de PLATA para ti! ğŸ¥ˆ ")
elif 3 <= sum(marcador) <= 4:
        print("\nMuy bien! Medalla de BRONCE para ti! ğŸ¥‰ ")
elif sum(marcador) <2:
        print("\nHey!! No te desanimes, ningÃºn deportista llegÃ³ a las olimpiadas sin antes haber entrenado duro! Seguro a la prÃ³xima te va mejor ğŸ‹ï¸â€â™€ï¸")
    
    
```
